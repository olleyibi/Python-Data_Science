## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the styles to Seaborn
sns.set()
# Import the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
# Load the country clusters data
data = pd.read_csv('C:/Users/iolley2/Desktop/DS Contd/Cluster Analysis/K-Means/Examples/Categorical.csv')
data.head()
```

Remove the duplicate index column from the dataset. 

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data_mapped = data.copy()
data_mapped['continent'] = data_mapped['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
data_mapped
```

## Select the features

```python
x = data_mapped.iloc[:,3:4]
x
```

## Clustering

Use 4 clusters initially.

```python
kmeans4 = KMeans(4)
kmeans3 = KMeans(3)
kmeans7 = KMeans(7)
kmeans8 = KMeans(8)
```

```python
kmeans4.fit(x)
kmeans3.fit(x)
kmeans7.fit(x)
kmeans8.fit(x)
```

## Clustering results

```python
identified_clusters4 = kmeans4.fit_predict(x)
identified_clusters3 = kmeans3.fit_predict(x)
identified_clusters7 = kmeans7.fit_predict(x)
identified_clusters8 = kmeans8.fit_predict(x)
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster4'] = identified_clusters4
data_with_clusters['Cluster3'] = identified_clusters3
data_with_clusters['Cluster7'] = identified_clusters7
data_with_clusters['Cluster8'] = identified_clusters8
```

## Plot the data

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_with_clusters['Cluster4'],cmap='rainbow')
plt.show()
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_with_clusters['Cluster3'],cmap='rainbow')
plt.show()
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_with_clusters['Cluster7'],cmap='rainbow')
plt.show()
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_with_clusters['Cluster8'],cmap='rainbow')
plt.show()
```
