# Train Test Split Example
## Import Libraries

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

## Generate/import data

```python
a = np.arange(1,101)
a
```

```python
b = np.arange(501,601)
b
```

```python
c = np.arange(501,601)
c
```

## Split data

```python
train_test_split(a,random_state =1)
```

```python
a_train,a_test = train_test_split(a,test_size = 0.2)
```

## Explore  the result

```python
a_train.shape, a_test.shape # check the shapes
```

```python
train_test_split(a,b,c,c,test_size = 0.2)
```
