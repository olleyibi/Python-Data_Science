# Preprocessing data

###### Import Libraries

```python
import pandas as pd
import numpy as np
```

###### Import Data

```python
raw_csv_data = pd.read_csv('C:/Users/iolley2/Desktop/DS Contd/Absenteeism Case/Absenteeism_data.csv')
raw_data = raw_csv_data.copy()
raw_data.head(10)
```

######  Drop the ‘ID’ column

```python
raw_data = raw_data.iloc[:,1:]
raw_data.head(10)
```

###### Group 'Reason for absence' column into 4 groups

```python
reason = []
dat = raw_data['Reason for Absence']
for i in range(raw_data.shape[0]):
    if dat[i] > 0 and dat[i] <= 14:
        reason.append('1')
    elif dat[i] > 14 and dat[i] <= 17:
        reason.append('2')
    elif dat[i] > 17 and dat[i] <= 21:
        reason.append('3')
    elif dat[i] > 21:
        reason.append('4')
    else:
        reason.append('0')
```

```python
raw_data['Reason for Absence'] = reason
raw_data.columns = raw_data.columns.str.replace('Reason for Absence','Reason')
raw_data.head(10)
```

###### One hot encode 'reason' column

```python
raw_data = pd.get_dummies(raw_data,
               columns = ['Reason'])
```

```python
raw_data = raw_data.drop(['Reason_0'], axis = 1)
```

```python
raw_data.columns
```

###### rearrange columns

```python
raw_data = pd.DataFrame(raw_data,columns = ['Reason_1', 'Reason_2', 'Reason_3',
       'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours'])
```

```python
raw_data
```

###### Create Checkpoints

```python
raw_data2 = raw_data.copy()
```

###### Continue

```python
raw_data2['Date'] = pd.to_datetime(raw_data2['Date'],format = '%d/%m/%Y')
raw_data2['Month Value'] = raw_data2['Date'].dt.month
```

```python
raw_data2['Month Value']
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()

raw_data2['Day of the Week'] = raw_data2['Date'].apply(date_to_weekday)
```

```python
raw_data2.head(5)
```

```python
raw_data2.columns
```

```python
col_reord = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']

raw_data2 = pd.DataFrame(raw_data2,columns = col_reord)
raw_data2.head(5)
```

```python
raw_data2['Education'] = raw_data2['Education'].map({1:0,2:1,3:1,4:1})
raw_data2['Education'].unique()
```
